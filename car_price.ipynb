```python
features = fdataset3.drop(['selling_price'], axis=1)
labels = fdataset3['selling_price']
features.head()
```

#Scaling

##Pour que les coefficients restent dans les normes et que le bias ne soit pas trop leve , on standardise ==> scaling

```python
# Normalisation: nrodou les valeurs entre un min et un max
#ma 7ajetnech biha since valeuret wadh7in w 9rab
from sklearn.preprocessing import PolynomialFeatures, MinMaxScaler
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.25, random_state=42) #optimization: stratify
```

```python
#standardisation
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(features)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import validation_curve
from sklearn.metrics import r2_score
from sklearn.metrics import f1_score

```

```python
from sklearn.tree import DecisionTreeRegressor
modelDTR = DecisionTreeRegressor()
modelDTR.fit( X_train, y_train)
modelDTR.score(X_test, y_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error 

y_preds1 = modelDTR.predict(X_train)
print(mean_absolute_error(y_preds1, y_train), mean_squared_error(y_preds1, y_train))
```

```python

y_preds2 = lr.predict(X_train)
print(mean_absolute_error(y_preds2, y_train), mean_squared_error(y_preds2, y_train))
```
